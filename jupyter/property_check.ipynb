```python
#財產審核
import pandas as pd
import numpy as np
pd.options.display.max_colwidth=0 #顯示全部字串
#匯入縣府財管系統盤點清冊
p = pd.read_excel(r'C:\Users\fhopecc\Google 雲端硬碟\教育基金108期中財抽\美崙國中\財產物品\美崙財產.xlsx')
i = pd.read_excel(r'C:\Users\fhopecc\Google 雲端硬碟\教育基金108期中財抽\美崙國中\財產物品\美崙物品.xlsx')
p.columns=["財產編號\分號\登錄號","財產名稱", "型式\品名", "單位", "數量", "殘值\累計折舊"
          ,"原始金額\帳面價值", "購置日期\領用日期", "使用年限\已用年限", "存放地點"
          ,"保管人員", "備註", "檢查情形"
          ]
p = p.dropna(thresh=4)
p = p[~(p["財產名稱"]=="財產名稱")]
p[['財產編號','登錄號']] = p['財產編號\分號\登錄號'].str.split('\n',expand=True)
p[['原始金額','帳面價值']] = p['原始金額\帳面價值'].str.split('\n',expand=True)
p[['殘值','累計折舊']] = p['殘值\累計折舊'].str.split('\n',expand=True)
p[['購置日期','領用日期']] = p['購置日期\領用日期'].str.split('\n',expand=True)
p[['使用年限','已用年限']] = p['使用年限\已用年限'].str.split('\n',expand=True)
p=p[['財產編號', '登錄號', '財產名稱', '型式\品名'
    , '單位', '存放地點', '保管人員', '備註'
    , '購置日期', '領用日期', '使用年限', '已用年限'
    , '原始金額', '帳面價值', '數量'
    , '殘值', '累計折舊'
    ]]
p['原始金額'] = p['原始金額'].str.replace(',','')
p['帳面價值'] = p['帳面價值'].str.replace(',','')
p=p.astype({'原始金額':'int32', '帳面價值':'int32'})
p['財產名稱']=p['財產名稱'].str.replace(' ', '')
p['財產名稱']=p['財產名稱'].str.replace('\n', '')

i.columns=["物品編號\分號\登錄號", "舊登錄號", "財產名稱"
          ,"型式品名", "單位", "數量", "總價", "購置日期\領用日期" 
          ,"使用年限", "已用年限", "存放地點", "保管人員", "備註", "檢查情形"
          ]
i = i.dropna(thresh=4)
i = i[~(i["財產名稱"]=="財產名稱")]
i[['物品編號', '登錄號']] = i['物品編號\分號\登錄號'].str.split('\n',expand=True)
i[['購置日期','領用日期']] = i['購置日期\領用日期'].str.split('\n',expand=True)
i=i[['物品編號', '登錄號', '財產名稱', '型式品名'
     , '單位', '存放地點', '保管人員', '備註'
     , '購置日期', '領用日期',  '數量', '總價'
     , '使用年限', '已用年限', '檢查情形', '舊登錄號' 
    ]]
i['財產名稱'] = i['財產名稱'].str.replace('\n','')
i['財產名稱']=i['財產名稱'].str.replace(' ', '')
i['備註'] = i['備註'].str.replace('\n','')
i['備註']=i['備註'].str.replace(' ', '')
i['型式品名'] = i['型式品名'].str.replace('\n','')
i['型式品名'] = i['型式品名'].str.replace(' ','')
i['總價'] = i['總價'].str.replace(',','')
i=i.astype({'總價':'float'})
```

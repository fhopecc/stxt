```python
import pandas as pd
import numpy as np
import re
pd.options.display.max_colwidth=0 #顯示全部字串
#清除機構營運紀錄
r = pd.DataFrame()
#excel
dhx = pd.read_excel(r'D:\gdrive\環保基金財抽\審核通知\第2次聲復\東和107年度營運紀錄.xls',sheet_name=None) 
dfx = pd.read_excel(r'D:\gdrive\環保基金財抽\審核通知\第2次聲復\東豐107年度營運紀錄.xls',sheet_name=None) 
#匯入東和
for k, v in dhx.items():
    if(re.search('申報', k)==None):
         r = pd.concat([r, v], ignore_index=True)
#匯入東豐
for k, v in dfx.items():
    if(re.search('申報', k)==None):
         r = pd.concat([r, v], ignore_index=True)
#清理雜訊資料
r.columns
r=r.dropna(subset=['單號'])

#加入計算欄位
r['清除機構簡稱'] = np.where(r[r'清除管編'].str.contains('東豐'),'東豐', '東和')

#r.shape
#r.head()
#匯入清查無營運紀錄卻有交付食用油清冊
#來源excel檔
tx = pd.read_excel(r'D:\gdrive\環保基金財抽\審核通知\附表5-未申報營運紀錄之清除機構-通知.xlsx'
                  ,sheet_name=[r'本縣已交付清除機構但未申報營運紀錄者']
                  ,skiprows=1
                  ,skipfooter=1
                  )
                
t=tx[r'本縣已交付清除機構但未申報營運紀錄者']
```

```python
t.columns
```

```python
t1=t.merge(r
          ,how='left'
          ,left_on=['清除機構','產源清冊名稱']
          ,right_on=['清除機構簡稱','委託單位名稱']          
          )

t1["最近申報日期"] = t1.groupby('系統序號')["清運日期"].transform('max')
t1.to_csv("waste_oil.csv")
```

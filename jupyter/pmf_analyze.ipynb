```python
#請購明細是支出分類最細
#排除退還保證金
ap1 = ap[(~ap['子目名稱'].str.contains('保證',na=False))&(~ap['總帳科目代碼'].str.contains('115Y',na=False))]
#排除分支計畫為經常門者
ap2 = ap1[~ap1['分支計畫名稱'].str.contains('各校經常門分支計畫',na=False)]
#排除摘要顯非工程者
ap3 = ap2[~ap2['摘要'].str.contains('樂器|鋼琴|工作坊|轉保固金|扣繳|繳庫|繳回|教材印刷|美術設計|弈宣設計|室外佈告欄設備|字幕機設備工程|內頁設計|電梯保養費|結餘款|美術班設計課|希望工程|美術設計', na=False)]

ap4 = ap3[ap3['摘要'].str.contains('工程|設計|監造',na=False)]

#判斷費用類形

ap4['費用類型']=np.select(
    [
        ap4['摘要'].str.contains('代扣所得稅',na=False),
        ap4['摘要'].str.contains('設計|監造|事務所',na=False),
        ap4['摘要'].str.contains('工程管理|督導|旅費|採購中心作業費',na=False),
        ap4['摘要'].str.contains('空氣污染防制費',na=False),
    ],
    [
        '其他',
        '委辦技服費',
        '工程管理費',
        '其他'
    ],
    default='工程款'
)

ap4[ap4['費用類型'].str.contains('工程管理費')][
    ['機關名稱', '年度', '請購單編號', '請購單明細序號' , '摘要', '請購金額'
    ,'總帳科目代碼', '子目', '子目名稱', '分支計畫代碼'
    ,'分支計畫名稱', '用途代碼'
    ]].sort_values('請購金額', ascending=False)
```
